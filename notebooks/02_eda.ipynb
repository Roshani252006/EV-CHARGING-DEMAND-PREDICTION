````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/ev_charging_patterns[1].csv')
df.groupby('Charger Type')['Energy Consumed (kWh)'].sum().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(7,7)
)

plt.title('Energy Consumption by Charger Type')
plt.ylabel('')
plt.show()
````

````python
import matplotlib.pyplot as plt

df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'])
df['Hour'] = df['Charging Start Time'].dt.hour
hourly_usage = df.groupby('Hour')['Energy Consumed (kWh)'].sum()

hourly_usage.plot(kind='bar', figsize=(10,5))
plt.title('Energy Consumption by Hour')
plt.xlabel('Hour')
plt.ylabel('Energy Consumed (kWh)')
plt.show()
````
